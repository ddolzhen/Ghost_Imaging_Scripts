```python
import random


import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from matplotlib.ticker import NullFormatter


import GI_simulations
```

```python
frames=GI_simulations.createFrames(10000,(100,100))
GI_simulations.addSmallSpecklesNegativeBinomial(frames,40,0.5)
frames_with_object=np.copy(frames)

mask=GI_simulations.definedMasks("square",frames_with_object)
GI_simulations.cutOutMask(frames_with_object,mask)
```

```python
fig, ax = plt.subplots(ncols=2,nrows=1, figsize=(30, 30))
frame_index=random.randrange(frames.shape[0])


_=ax[0].imshow(frames[frame_index,:,:])
_=ax[1].imshow(frames_with_object[frame_index,:,:])
```

```python
ghost=GI_simulations.ghostPixel(frames,frames_with_object)
```

```python


#Show the ghost
fig, ax = plt.subplots(ncols=1,nrows=1, figsize=(20, 20))
_=ax.imshow(ghost)
fig.colorbar(_)

```
